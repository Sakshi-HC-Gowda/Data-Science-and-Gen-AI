```python
# probability Distributions

# Roll a dice

# 1,2,3,4,5,6

import random

results = []

for i in range(10):
  roll = random.randint(1,6)
  results.append(roll)

print(results)
```

```python
from collections import Counter

frequency = Counter(results)

print(frequency)
```

Probability = Frequency/Total

```python
total = len(results)

for number,count in frequency.items():
  prob = count/total
  print(f'The probability of {number} is {prob}')
```

**Binomial Distributions**

- Fixed number of trials
- Only two possible outcomes

```python
# Binomial Distribution

from scipy.stats import binom
import matplotlib.pyplot as plt

n = 10    # number of tosses
p = 0.5   # probability of head

x = range(0,11)

probs = binom.pmf(x,n,p)

plt.bar(x,probs)
plt.xlabel("Number of Heads")
plt.ylabel("probability")
plt.show()
```

```python
# probability of getting a head is 0.5
# toss the coin 10 times -> probability of getting n number of heads
```

**Hypothesis Testing**

```python
# the average marks of the students are 70 -> this process is called as Hypothesis Testing

# Null hypothesis (H0) - The default assumption


H0 = "Average marks - 70"
print(H0)
```

```python
# Alternative Hypothesis(H1) - AH is what we are trying to prove

# Maybe the average marks are not 70

H1 = "Average Marks is NOT equal to 70"
print(H1)
```

```python
# sample data

marks = [68,72,75,71,69,74,73]
average = sum(marks)/len(marks)
print(average)
```

```python
# Hypothesis Test (One sample t-Test)

from scipy.stats import ttest_1samp

marks = [68,72,75,71,69,74,73]

t_stat,p_value = ttest_1samp(marks,70)

print(t_stat)
print(p_value)
```

```python
# H0 is likely to be Accepted/Rejected
```

```python
# Independent t-Test

# class A,B

A = [60,65,62,70,68]
B = [75,80,82,78,79]

from scipy.stats import ttest_ind

t,p = ttest_ind(A,B)
print(p)
```

```python
# Paired t-Test

before = [55,60,58,62,57]
after = [68,72,70,75,69]

from scipy.stats import ttest_rel

t,p = ttest_rel(before,after)

print(p)
```

```python
# correlation

# r - close to 1 - Strong positive relationship
# r - close to 0 - little or no linear relationship
# r - clsoe to -1 - Strong negative relationship


hours = [2,4,6,8,10]
marks = [50,60,70,80,90]

from scipy.stats import pearsonr

r,p = pearsonr(hours,marks)

print(r)
print(p)
```

```python
# Shapiro-wilk Test - if p > 0.05 : data is approximately normal, if p < 0.05 : data is not normal

from scipy.stats import shapiro

marks = [68,72,75,71,69,74,73]

stat, p = shapiro(marks)

print(p)
```
