```python
class ELECTRONICS():
  def ECE(self):
    print('This is ECE')
  def EEE(self):
    print('This is EEE')

obj = ELECTRONICS()
obj.EEE()
```

- Create a class and execute all the arithemetic operations using different functions

```python
# Inheritance

class PARENT():
  def properties(self):
    print('We have properties')

class CHILD(PARENT):
  def bike(self):
    print('I own a bike')

obj2 = CHILD()
obj2.properties()
```

```python
# Multi level Inheritance

class GRANDPARENT():
  def strength(self):
    print('We stay strong')

class PARENT1(GRANDPARENT):
  def assets(self):
    print('We have assets')

class CHILD2(PARENT1):
  def fun(self):
    print('We have a lot of fun')


obj3 = CHILD2()
obj3.strength()
```

```python
class PARENT2():
  def add(self):
    pass

class PARENT3():
  def sub(self):
    pass

class CHILD3(PARENT2,PARENT3):
  def op(self):
    pass


obj4 = CHILD3()
obj4.add()
```
