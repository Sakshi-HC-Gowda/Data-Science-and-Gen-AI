**String**

```python
s = "python programming"
#capitalize() → First letter uppercase
print(s.capitalize())
```

```python
#title() → Every word's first letter uppercase
print(s.title())
```

```python
#upper() → All uppercase
print(s.upper())
#lower() → All lowercase
print(s.lower())
```

```python
#swapcase() → Changes upper ↔ lower
print(s.swapcase())
#strip() → Removes starting and ending spaces
print(s.strip())
```

```python
#replace() → Replaces text
print(s.replace("python", "java"))
```

```python
#split() → String to list
print(s.split())
```

```python
#join() → Joins elements into string
print("-".join(s.split()))
```

```python
#find() → Finds index
print(s.find("programming"))
```

```python
#count() → Counts occurrences
print(s.count("m"))
```

```python
#startswith() → Checks starting text
print(s.startswith("python"))
```

```python
#endswith() → Checks ending text
print(s.endswith("programming"))
```

**List**

```python
a = [30, 10, 20]
a.append(40)
print(a)
```

```python
a.extend([50, 60])
print(a)
```

```python
a.insert(1,15)
print(a)
```

```python
a.remove(20)
print(a)
```

```python
a.pop()
print(a)
```

```python
b=a.copy()
print(b)
```

```python
a.sort()
print(a)
```

```python
a.reverse()
print(a)
```

```python
a.clear()
print(a)
```

**Tuple**

```python
t = (10, 20, 30, 20)

print(t.count(20))
print(t.index(30))
```

**Set**

```python
a = {1, 2, 3}
b = {3, 4, 5}

a.add(6)
print(a)
```

```python
a.update([7, 8])
print(a)
```

```python
a.remove(8)
print(a)
```

```python
a.discard(10)
print(a)
```

```python
print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
print(a.symmetric_difference(b))
print(a.pop())
```

**Dictionary**

```python
d={"name":"Sakshi","age":20}
print(d.keys())
print(d.values())
print(d.items())
print(d.get("name"))
```

```python
d.update({"city": "Udupi"})
print(d)
```

```python
d.pop("age")
print(d)
```

```python
d.setdefault("country", "India")
print(d)
```

```python
d.popitem()
print(d)
```

```python
d.clear()
print(d)
```
