```python
if 'Unnamed: 32' in cancer.columns:
    cancer.drop("Unnamed: 32", axis=1, inplace=True)
    print("Column 'Unnamed: 32' dropped successfully.")
else:
    print("Column 'Unnamed: 32' not found in DataFrame.")
```

```python
cancer.drop_duplicates(inplace=True)
```

```python
cancer["diagnosis"] = cancer["diagnosis"].replace({
    "M": "Malignant",
    "B": "Benign"
})
```

```python
cancer["diagnosis"].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="diagnosis", data=cancer)

plt.title("Distribution of Breast Cancer Diagnosis")
plt.xlabel("Diagnosis")
plt.ylabel("Number of Patients")

plt.show()
```

```python
sns.histplot(
    data=cancer,
    x="radius_mean",
    hue="diagnosis",
    kde=True
)

plt.title("Tumor Radius Distribution by Diagnosis")
plt.xlabel("Mean Radius")

plt.show()
```

```python
sns.scatterplot(
    x="radius_mean",
    y="texture_mean",
    hue="diagnosis",
    data=cancer
)

plt.title("Tumor Radius vs Texture")
plt.xlabel("Mean Radius")
plt.ylabel("Mean Texture")

plt.show()
```

```python
sns.boxplot(
    x="diagnosis",
    y="area_mean",
    data=cancer
)

plt.title("Tumor Area by Diagnosis")
plt.xlabel("Diagnosis")
plt.ylabel("Mean Area")

plt.show()
```

```python
features = cancer[[
    "radius_mean",
    "texture_mean",
    "perimeter_mean",
    "area_mean",
    "smoothness_mean",
    "compactness_mean",
    "concavity_mean"
]]

plt.figure(figsize=(10, 6))

sns.heatmap(
    features.corr(),
    annot=True,
    fmt=".2f"
)

plt.title("Correlation Between Tumor Features")

plt.show()
```
