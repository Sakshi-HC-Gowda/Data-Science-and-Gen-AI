```python
df_mean = df.copy()
```

```python
mean_value = df_mean["NumberOfDependents"].mean()

print("Mean value:", mean_value)
```

```python
df_mean["NumberOfDependents"] = df_mean["NumberOfDependents"].fillna(mean_value)
```

```python
mean_after_mean = df_mean["NumberOfDependents"].mean()
print("Mean before filling:",mean_before)
print("Mean after Mean filling:",mean_after_mean)
```

```python
df_median = df.copy()
```

```python
median_value=df_median["NumberOfDependents"].median()
print("Median value:", median_value)
```

```python
df_median["NumberOfDependents"] = df_median["NumberOfDependents"].fillna(median_value)
```

```python
mean_after_median = df_median["NumberOfDependents"].mean()
print("Mean before filling:", mean_before)
print("Mean after Median filling:", mean_after_median)
```

```python
df_mode = df.copy()
```

```python
mode_value = df_mode["NumberOfDependents"].mode()[0]

print("Mode value:", mode_value)
```

```python
df_mode["NumberOfDependents"] = df_mode["NumberOfDependents"].fillna(mode_value)
```

```python
mean_after_mode = df_mode["NumberOfDependents"].mean()
print("Mean before filling:", mean_before)
print("Mean after Mode filling:", mean_after_mode)
```

**TASK 2: CREDIT CARD DATA VISUALIZATION**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(x="Good_Bad",data=df)
plt.title("Good vs Bad Credit Distribution")
plt.xlabel("Credit Status")
plt.ylabel("Number of Customers")
plt.show()
```

```python
sns.histplot(df["age"], bins=30, kde=True)

plt.title("Age Distribution of Customers")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
sns.histplot(df["MonthlyIncome"], bins=50)

plt.title("Distribution of Monthly Income")
plt.xlabel("Monthly Income")
plt.ylabel("Frequency")

plt.show()
```

```python
sns.boxplot(x=df["MonthlyIncome"])

plt.title("Monthly Income and Outliers")

plt.show()
```

```python
sns.countplot(x="Gender", hue="Good_Bad", data=df)

plt.title("Credit Status by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

```python
sns.countplot(x="Region", hue="Good_Bad", data=df)

plt.title("Credit Status by Region")
plt.xlabel("Region")
plt.ylabel("Count")

plt.xticks(rotation=45)

plt.show()
```

```python
sns.scatterplot(
    x="DebtRatio",
    y="MonthlyIncome",
    hue="Good_Bad",
    data=df
)

plt.title("Debt Ratio vs Monthly Income")

plt.show()
```

```python
numeric_data = df.select_dtypes(include="number")

plt.figure(figsize=(12, 8))

sns.heatmap(
    numeric_data.corr(),
    annot=True,
    fmt=".2f"
)

plt.title("Correlation Heatmap of Credit Data")

plt.show()
```
