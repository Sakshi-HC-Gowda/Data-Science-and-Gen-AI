```python
#Create a class and execute all the arithmetic operations using different functions

class Arithmetic:

    def add(self, a, b):
        print("Addition =", a + b)

    def sub(self, a, b):
        print("Subtraction =", a - b)

    def mul(self, a, b):
        print("Multiplication =", a * b)

    def div(self, a, b):
        print("Division =", a / b)


obj = Arithmetic()

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

obj.add(a, b)
obj.sub(a, b)
obj.mul(a, b)
obj.div(a, b)
```

```python
#SINGLE LEVEL INHERITANCE
class PARENT():
  def properties(self):
    print('We have properties')
class CHILD(PARENT):
  def bile(self):
    print('I own a bike')
obj2=CHILD()
obj2.properties()
```

```python
#MULTI LEVEL INHERITANCE
class GRANDPARENT():
  def strength(self):
    print('We stay strong')
class PARENT1(GRANDPARENT):
  def assests(self):
    print('We have assests')

class CHILD2(PARENT1):
  def fun(self):
    print('We have a lot of fun')
obj3=CHILD2()
obj3.strength()
```

```python

class PARENT2():
  def add(self):
    pass
class PARENT3():
  def sub(self):
    pass

class CHILD3(PARENT2,PARENT3):
  def op(self):
    pass

obj4=CHILD3()
obj4.add()
```

```python
# HIERARCHICAL INHERITANCE
class PARENT():
    def properties(self):
        print("We have properties")

class CHILD1(PARENT):
    def bike(self):
        print("I own a bike")

class CHILD2(PARENT):
    def car(self):
        print("I own a car")

obj1 = CHILD1()
obj1.properties()
obj1.bike()

obj2 = CHILD2()
obj2.properties()
obj2.car()
```

```python
# HYBRID INHERITANCE
class GRANDPARENT():
    def strength(self):
        print("We stay strong")

class PARENT1(GRANDPARENT):
    def house(self):
        print("We have a house")

class PARENT2(GRANDPARENT):
    def land(self):
        print("We have land")

class CHILD(PARENT1, PARENT2):
    def bike(self):
        print("I own a bike")

obj = CHILD()
obj.strength()
obj.house()
obj.land()
obj.bike()
```
