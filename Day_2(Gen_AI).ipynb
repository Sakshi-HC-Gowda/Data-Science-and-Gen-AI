```python
import matplotlib.pyplot as plt
```

```python
l1 = [2,5,7,3,8,6]
l2 = [1,3,4,2,6,8]

plt.scatter(x=l1,y=l2)
plt.show()
```

```python
plt.figure(figsize=(5,5))  #w,h

l1 = [2,5,7,3,8,6]
l2 = [1,3,4,2,6,8]

plt.scatter(x=l1,y=l2,marker='*',color='g')
plt.show()
```

```python
plt.figure(figsize=(5,5))  #w,h

l1 = [2,5,7,3,8,6]
l2 = [1,3,4,2,6,8]

plt.scatter(x=l1,y=l2,marker='*',color='g')
plt.title('Scatter_Plot')
plt.xlabel('List1')
plt.ylabel('List2')
plt.show()
```

**Line Chart**

```python
plt.figure(figsize=(8,6))  #w,h

l1 = [2,5,7,3,8,6]
l2 = [1,3,4,2,6,8]

plt.plot(l1,l2,marker='*',color='g')
plt.show()
```

```python
Kannada = [2,5,7,8,9]
Tulu = [5,7,9,2,1]
English = [9,4,2,7,1]

plt.plot(Kannada,Tulu,label='KT')
plt.plot(Kannada,English,label='KE')
plt.legend(loc=1,fontsize='x-large')

plt.show()
```

**SubPlots**

```python
plt.figure(figsize=(5,9))

Kannada = [2,5,7,8,9]
Tulu = [5,7,9,2,1]
English = [9,4,2,7,1]

plt.subplot(2,1,1)
plt.title('KT')
plt.plot(Kannada,Tulu,color='g',marker='*')

plt.subplot(2,1,2)
plt.title('KE')
plt.plot(Kannada,English,color='r',marker='*')

plt.show()
```

**Bar charts**

```python
plt.figure(figsize=(8,3))

company = ['Infosys','Intel','Deloitte','Zoho','Samsung']
revenue = [1500,1800,2200,1500,3000]

plt.bar(company,revenue,color=['r','g','y','black','b'])

plt.show()
```

```python
plt.figure(figsize=(8,3))

company = ['Infosys','Intel','Deloitte','Zoho','Samsung']
revenue = [1500,1800,2200,1500,3000]

plt.barh(company,revenue,color=['r','g','y','black','b'])

plt.show()
```

**Pie Chart**

```python
plt.figure(figsize=(8,3))

company = ['Infosys','Intel','Deloitte','Zoho','Samsung']
revenue = [1500,1800,2200,1500,3000]

plt.pie(revenue,labels=company)

plt.show()
```

```python
plt.figure(figsize=(8,3))

company = ['Infosys','Intel','Deloitte','Zoho','Samsung']
revenue = [1500,1800,2200,1500,3000]

plt.pie(revenue,labels=company,shadow=True)

plt.show()
```

```python
plt.figure(figsize=(8,3))

company = ['Infosys','Intel','Deloitte','Zoho','Samsung']
revenue = [1532,1868,2200,1500,3000]

plt.pie(revenue,labels=company,autopct='%.2f')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing


data = fetch_california_housing(as_frame=True).frame


correlation_matrix = data.corr()


plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix of California Housing Dataset")
plt.show()


sns.pairplot(data, diag_kind='kde')
plt.suptitle("Pairplot of California Housing Features", y=1.07)
plt.show()
```
